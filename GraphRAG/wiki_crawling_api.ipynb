# Get Wiki texts through API

```python
import requests

def get_wiki_text(title):
    url = "https://en.wikipedia.org/w/api.php"
    params = {
        "action": "query",
        "format": "json",
        "prop": "extracts",
        "explaintext": True,
        "titles": title
    }

    headers = {
        "User-Agent": "NTU-GraphRAG-Assignment/1.0 (student project; )"
    }

    res = requests.get(url, params=params, headers=headers)
    print("STATUS:", res.status_code) # debug line
    print("TEXT PREVIEW:", res.text[:20])  

    data = res.json()
    pages = data["query"]["pages"]

    return next(iter(pages.values())).get("extract", "")

topics = [
    "Marie Curie",
    "Radium",
    "Polonium",
    "Radioactive decay",
    "X-ray",
    "Medical imaging",
    "Nuclear medicine",
    "Radiation therapy"
]

with open("curie_things.txt", "w", encoding="utf-8") as f:
    for t in topics:
        f.write(f"# {t}\n")
        f.write(get_wiki_text(t))
        f.write("\n\n")
```
